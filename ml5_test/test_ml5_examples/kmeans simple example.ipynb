```python
from ml5_test import ml5_kmeans
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
test = ml5_kmeans.Kmeans()
```

```python
img = cv2.imread('world.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
resize_size = (100, 100)
img_rgb = cv2.resize(img_rgb, resize_size)
```

```python
plt.imshow(img_rgb)
```

```python
x, y, channel = img_rgb.shape
x, y, channel
```

```python
for i in range(x):
    for j in range(y):
        r, g, b = img_rgb[i,j]
        test.data.append({'r':r, 'g':g, 'b':b})
```

```python
test.initialize()
```

```python
output_img = []
for c in test.data_label:
    output_img.append(c)
output_img = np.array(output_img).reshape(resize_size)
```
